```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np
from itertools import cycle, islice
import matplotlib.pyplot as plt
from pandas.plotting import parallel_coordinates

%matplotlib inline
```

```python
data = pd.read_csv("/content/drive/MyDrive/academic/l4s2/big data/d2.csv")
data.head()
```

```python
features = ['District', 'Area', 'Price', 'Brand', 'Model', 
        'Trim/Edition','ModelYear', 'Condition', 'Transmission', 'BodyType','FuelType','EngineCapacity','Millage']
```

```python
select_df = data[features]
```

```python
select_df.columns
```

```python
select_df
```

```python
select_df['Price'] = select_df['Price'].astype(float)
select_df['ModelYear'] = select_df['ModelYear'].astype(int)
# select_df['EngineCapacity'] = select_df['EngineCapacity'].astype(str)
# select_df["EngineCapacity"] = df.EngineCapacity.apply(lambda x: float(x.replace("/cc","").replace("/cc","")))
select_df['EngineCapacity'] = select_df['EngineCapacity'].str.replace(',', '').str.replace(' cc', '').astype(float)
select_df['Millage'] = select_df['Millage'].str.replace(',', '').str.replace(' km', '').astype(float)
# select_df['EngineCapacity'] = select_df['EngineCapacity'].astype(float)
print(select_df)
```

```python
select_df.columns
```

```python
categorical_cols = ['District', 'Area', 'Brand', 'Model', 'Trim/Edition', 'Condition','Transmission','BodyType','FuelType']

encoded = pd.get_dummies(select_df, columns=categorical_cols, drop_first=True)
```

```python
select_df = pd.concat([encoded], axis=1)
```

```python
select_df.head()
```

```python
X = StandardScaler().fit_transform(select_df)
X
```

```python
kmeans = KMeans(n_clusters=3)
model = kmeans.fit(X)
print("model\n", model)
```

```python
centers = model.cluster_centers_
centers
```

```python
def pd_centers(featuresUsed, centers):
	colNames = list(featuresUsed)
	colNames.append('prediction')

	# Zip with a column called 'prediction' (index)
	Z = [np.append(A, index) for index, A in enumerate(centers)]

	# Convert to pandas data frame for plotting
	P = pd.DataFrame(Z, columns=colNames)
	P['prediction'] = P['prediction'].astype(int)
	return P
```

```python
# Function that creates Parallel Plots

def parallel_plot(data):
	my_colors = list(islice(cycle(['b', 'r', 'g', 'y', 'k']), None, len(data)))
	plt.figure(figsize=(15,8)).gca().axes.set_ylim([-3,+3])
	parallel_coordinates(data, 'prediction', color = my_colors, marker='o')
```

```python
P = pd_centers(list(select_df.columns), centers)
P
```
